```python
wb = xw.Book()
sht = wb.sheets('Sheet2')
```

```python
for n in [2,3,4,5,6,7,8]:
    wb.sheets.add("n={}".format(n))
    sht = wb.sheets("n={}".format(n))
    hh = h_gen(n)
    j = 0
    for h in hh:
        w = permutations([i for i in range(1, n+1)])
        print("h: {}".format(h))
        QQ = []
        for ww in w:
            #print("    perm: {}".format(ww))
            qq = PRSK(ww, h)[0]
            #print("    P: {}".format(qq))
            QQ.append(qq)

        PP = []
        for P in QQ:
            if P in PP:
                continue
            else:
                PP.append(P)

    #print(len(PP))

        i = 1
        #print("Pie")
        columns=["h", "P", "T", "content", "shape"]
        df = pd.DataFrame([], columns=columns)
        for P in PP:
            #print("    P: {}".format(P))
            T = transpose(pie(transpose(P)))
            df.loc[i, "h"] = "{}".format(h)
            df.loc[i, "P"] = "{}".format(P)
            df.loc[i, "T"] = "{}".format(T)
            df.loc[i, "content"] = "{}".format(content(T))
            df.loc[i, "shape"] = "{}".format(shape(T))
            i = i+1

        df = df.sort_values(by=["content", "shape"])

        sht.range((1,7*j+1)).value = df
        j = j + 1
```
